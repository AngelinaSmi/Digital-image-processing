```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

Загрузим и выведем изображение

```python
image = cv.imread('./lb3.png')
```

```python
plt.imshow(image[:,:,:])
```

Берем изображение в градациях серого для наилучшего отображения распределения яркости

```python
img_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

Применим медианный фильтр/соль и перец для убирания шума

```python
denoised = cv.medianBlur(img_rgb, 5)
plt.imshow(denoised)
```

Применяем негатив

```python
negative = 255 - denoised
plt.imshow(negative)
```

Выведем все этапы улучшения изображения в одну линию

```python
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Исходное")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Медианный")
plt.imshow(denoised)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Негатив")
plt.imshow(negative)
plt.axis("off")

plt.show()
```
