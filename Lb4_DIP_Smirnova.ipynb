Задание: улучшить изображение для пеервода в google и яндекс переводчиках.

Загрузим необходимые библиотеки

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

Загрузим и выведем изображение

```python
image = cv.imread('./lb4.jpg')
plt.imshow(image)
```

Преобразуем изображение в градации серого

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
image = gray
```

В обработке изображений часто используется процедура пороговой бинаризации -- разбиения изображения на две области, одна из которых содержит все пиксели со значением ниже некоторого порога, а другая содержи все пиксели со значением выше этого порога. Пороговое значение - threshold.

```python
threshold = 110
ret, thresh1 = cv.threshold(image, threshold, 255, cv.THRESH_BINARY)
plt.figure(figsize=(70, 46))
plt.subplot(1, 3, 1)
plt.imshow(thresh1, 'gray')
plt.show()
```
